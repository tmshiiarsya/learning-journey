```python
print("---Mulai---\n")

jumlah_ganjil = 0
jumlah_genap = 0

with open('data.txt', 'r') as f:

    for baris in f:
        angka = int(baris.strip()) # Membersihkan 'enter' (\n)

        if angka % 2 == 0:
            status = "Genap"
            jumlah_genap += 1
        else:
            status = "Ganjil"
            jumlah_ganjil += 1
        
        print(f"Angka {angka} adalah {status}")

    print("-----")
    print(f"\nJumlah angka ganjil :{jumlah_ganjil}")
    print(f"Jumlah angka genap: {jumlah_genap}")

print("\n---Selesai---")
```
